#### Business data was subset with the following steps:

1. open businesses only
2. filter out businesses that have 'Other' category
3. falls under 'restaurant' or 'food' or both
4. Las Vegas, Phoenix, Scottsdale, Charlotte, and Toronto

We can use the remaining business_id to filter out review and user data.

```python
import pandas as pd
import numpy as np
```

```python
review = pd.read_csv('review.csv', index_col = 'review_id')
review.shape
```

```python
b = pd.read_csv('business_r_and_f.csv', index_col='business_id')
b.shape
```

```python
b.head()
```

```python
# Use business_id in business data to subset

r = review[review['business_id'].isin(b.index)]
r.shape
```

```python
user = pd.read_csv('user.csv')
user.head()
```

```python
user.shape
```

```python
user.set_index('user_id', inplace=True)
user.head()
```

```python
# Reorder columns
col = ['name', 'review_count', 'average_stars', 'yelping_since', 'friends', 'fans', 'elite',
       'cool', 'funny', 'useful',
       'compliment_cool', 'compliment_cute', 'compliment_funny', 'compliment_hot', 'compliment_list',
       'compliment_more', 'compliment_note', 'compliment_photos', 'compliment_plain', 'compliment_profile',
       'compliment_writer']
```

```python
user = user[col]
```

```python
print(review.user_id.nunique())
print(r.user_id.nunique())

print(review.business_id.nunique())
print(r.business_id.nunique())
```

```python
u = user[user.index.isin(r.user_id.unique())]
u.shape
```

```python
u.head(50)
```

```python
u.describe()
```

```python
# create new column count of friends
u['friend_count'] = np.where(u['friends'] == 'None', 0, u['friends'].apply(lambda x: str(x).count(',') + 1))
```

```python
u.head(50)
```

```python
u[(u['review_count'] == 0) & (u['friend_count'] == 0) & (u['fans'] == 0)]
```

```python
# save subset to csv

r.to_csv('review_r_and_f')
u.to_csv('user_r_and_f')
```
