#### Remove punctuations

```python
import re
re.sub(r'[^\w\s]', '', 'woo hoo......had a fabulous experience with hog car tours today! these scooters were super awesome!!')
```

```python
df_lv_5['text_clean'] = df_lv_5['text_clean'].apply(lambda x: re.sub(r'[^\w\s]', '', x))
df_lv_5.head()
```

#### Remove stopwords, digits

```python
stop = set(stopwords.words('english'))
#punct = set(string.punctuation)
#lemma = WordNetLemmatizer()

df_lv_5['text_clean'] = df_lv_5['text_clean'].apply(lambda x: " ".join([i for i in x.split() if i not in stop]))
#df_lv_5['text_clean'] = df_lv_5['text_clean'].apply(lambda x: " ".join([i for i in x.split() if i not in punct]))
df_lv_5['text_clean'] = df_lv_5['text_clean'].apply(lambda x: " ".join([i for i in x.split() if not i.isdigit()]))
```
